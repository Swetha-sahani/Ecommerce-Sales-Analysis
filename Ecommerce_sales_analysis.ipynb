# E-Commerce Sales Analysis Using Python

## Project Objective

Analyze e-commerce sales data to identify sales trends, profitable categories, high-performing regions, and loss-making products using Python and data visualization techniques.



1. Dataset Loading
Import Required Libraries
Load Dataset
Display First 5 Rows

```python
import pandas as pd

df = pd.read_csv('/content/SampleSuperstore.csv')

df.head()
```

2. Data Overview
Dataset Shape
Dataset Columns
Missing Values Analysis
Statistical Summary

```python
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())
```

3. KPI Analysis
Total Sales
Total Profit
Total Quantity Sold
Average Discount
Key Observations

```python
print("Total Sales:", df['Sales'].sum())

print("Total Profit:", df['Profit'].sum())

print("Total Quantity Sold:", df['Quantity'].sum())

print("Average Discount:", df['Discount'].mean()*100)
```

4. Category Analysis
Sales by Category
Visualization
Insights
Profit by Category

Insights

```python
category_sales = df.groupby('Category')['Sales'].sum()

print(category_sales)
```

```python
import matplotlib.pyplot as plt

category_sales.plot(kind='bar')

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
top_states = (
    df.groupby('State')['Sales']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(top_states)
```

6. State Analysis
Top 10 States by Sales
Visualization
Insights

```python
top_states = (
    df.groupby('State')['Sales']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(top_states)
```

```python
import matplotlib.pyplot as plt

top_states.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Sales')

plt.show()
```

```python
top_profit_states = (
    df.groupby('State')['Profit']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(top_profit_states)

top_profit_states.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 States by Profit')
plt.xlabel('State')
plt.ylabel('Profit')

plt.show()
```

### Insights

- California generated the highest profit among all states.
- States with high sales generally showed higher profitability.
- Profit distribution varied significantly across states, indicating regional differences in business performance.

8. Discount Analysis
Discount vs Profit Relationship
Visualization
Insights

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(
    df['Discount'],
    df['Profit'],
    alpha=0.5
)

plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Discount vs Profit")

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2, figsize=(15,10))

# Sales by Category
df.groupby('Category')['Sales'].sum().plot(
    kind='bar',
    ax=ax[0,0],
    title='Sales by Category'
)

# Profit by Category
df.groupby('Category')['Profit'].sum().plot(
    kind='bar',
    ax=ax[0,1],
    title='Profit by Category'
)

# Region Sales
df.groupby('Region')['Sales'].sum().plot(
    kind='pie',
    autopct='%1.1f%%',
    ax=ax[1,0]
)

ax[1,0].set_ylabel("")

# Top States
df.groupby('State')['Sales'].sum().sort_values(
    ascending=False
).head(10).plot(
    kind='bar',
    ax=ax[1,1],
    title='Top States'
)

plt.tight_layout()
plt.show()
```

# Business Insights

## Key Findings

1. Technology generated the highest sales and profit.

2. Furniture showed low profitability despite high sales.

3. West region contributed the highest revenue.

4. California was the top-performing state.

5. Tables were the biggest loss-making product.

6. High discounts negatively impacted profitability.

# Conclusion

The analysis revealed that Technology is the most profitable category, while Furniture generates high sales but relatively low profit. The West region and California contribute significantly to revenue. Loss-making products such as Tables should be reviewed, and discount strategies should be optimized to improve overall profitability.
